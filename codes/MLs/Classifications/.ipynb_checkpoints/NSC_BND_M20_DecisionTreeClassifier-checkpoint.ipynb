```python
import pymongo as mg
import pandas as pd
```

```python
client = mg.MongoClient(host= 'mongodb://localhost:27017')
```

```python
database = client['db_NHIS']
collection = database['merged_collection6']
```

```python
cursor = collection.find({})
list_BND_M20 = list(cursor)
```

```python
df_BND_M20_merge = pd.DataFrame(list_BND_M20)
df_BND_M20_merge[:2]
```

```python
df_BND_M20_merge.columns
```

```python
df_BND_M20_merge['SICK_SYM1'].value_counts()
```

```python
df_BND_M20 = df_BND_M20_merge[['SICK_SYM1', 'MDCARE_DD_CNT','ED_RC_TOT_AMT','EDC_SBA', 'ED_RC_TOT_AMT','MCARE_SUBJ_CD']]
df_BND_M20.isnull().sum() 
```

```python
len('F_'), len('J209'), 'J209'[:3], str('T18')[:3]
```

```python
def convertSICK_SYM(sick_sym) : 
    if len(sick_sym) < 3:
        return None
    else :
        return sick_sym[:3]
```

```python
convertSICK_SYM('J209'), convertSICK_SYM('F_')
```

```python
df_BND_M20['SICK_SYM1_3'] = df_BND_M20['SICK_SYM1'].apply(convertSICK_SYM)
```

```python
df_BND_M20['SICK_SYM1_3'].value_counts()
```

```python
df_BND_M20_drop = df_BND_M20.dropna()
df_BND_M20_drop.isnull().sum()
```

```python
df_BND_M20_drop[:2]
```

```python
target_list = df_BND_M20_drop['SICK_SYM1_3'].value_counts().index.to_list()
```

```python
target_list.index('L02'), target_list.index('J03')
```

```python
def applyIndexNumber(sick_sym_3) :
    indexNumber = target_list.index(sick_sym_3)
    return indexNumber
```

```python
df_BND_M20_drop['SICK_SYM1_3_target'] = df_BND_M20_drop['SICK_SYM1_3'].apply(applyIndexNumber)
df_BND_M20_drop[:2]
```

### 정형화

```python
from sklearn.model_selection import train_test_split
target = df_BND_M20_drop['SICK_SYM1_3_target']
features = df_BND_M20_drop[['MDCARE_DD_CNT','ED_RC_TOT_AMT','EDC_SBA','ED_RC_TOT_AMT','MCARE_SUBJ_CD']]
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=111)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

```python
target_train.shape, features_train.shape
```

### 학습

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(features_train,target_train)
```

```python
model.coef_, model.intercept_
```

```python
df_BND_M20_drop[250:255]
```

```python
model.predict(features_train[250:255])
```

```python
model.predict_proba(features_train[250:255])
```

### 평가

```python
target_train_predict = model.predict(features_train)
target_train_predict.shape  #target_train.shape과 동일
```

```python
from sklearn.metrics import accuracy_score #정확도
```

```python
accuracy_score(target_train, target_train_predict) #교내 시험
```

```python
target_test_predict = model.predict(features_test)
target_test_predict.shape #target_test.shape과 동일
```

```python
accuracy_score(target_test, target_test_predict)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target_train, target_train_predict))
```

```python
print(classification_report(target_test, target_test_predict))
```
